```python
import pickle
import os
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.patches as mpatches
```

```python
def calculate_macro_metrics(conf_matrix):
    num_classes = conf_matrix.shape[0]
    sensitivities = []
    specificities = []
    
    for i in range(num_classes):
        TP = conf_matrix[i, i]
        FN = np.sum(conf_matrix[i, :]) - TP
        FP = np.sum(conf_matrix[:, i]) - TP
        TN = np.sum(conf_matrix) - (TP + FN + FP)
        
        sensitivity_i = TP / (TP + FN)
        specificity_i = TN / (TN + FP)
        
        sensitivities.append(sensitivity_i)
        specificities.append(specificity_i)
    
    macro_sensitivity = np.mean(sensitivities)
    macro_specificity = np.mean(specificities)
    
    return macro_sensitivity, macro_specificity
```

```python
def sort_dict_by_values(d):
    sorted_dict = {k: v for k, v in sorted(d.items(), key=lambda item: item[1], reverse=True)}
    return sorted_dict
```

```python
# Model wise results

ba_results = {}
acc_results = {}
macro_preci_results = {}
macro_sensitivity_results = {}
macro_specificity_results = {}
macro_f1_score_results = {}

# -------------------------- ROCKET Results --------------------------
model = 'ROCKET'
ROCKET_model_evals = ROCKET_evals['additional_metrics']

# balanced accuracy
ba_results[model] = ROCKET_model_evals['balanced_accuracy']

# accuracy
acc_results[model] = ROCKET_model_evals['model_accuracy']

# macro precision
macro_preci_results[model] = ROCKET_model_evals['macro_precision']

# macro sensitivity and & specificity
macro_sensitivity_results[model], macro_specificity_results[model] = calculate_macro_metrics(ROCKET_evals['confusion_matirx'])

# macro f1-score
macro_f1_score_results[model] = ROCKET_model_evals['macro_f1-score']

# -------------------------- ConvTran Results --------------------------
model = 'ConvTran'
ConvTran_model_evals = ConvTran_evals['additional_metrics']

# balanced accuracy
ba_results[model] = ConvTran_model_evals['balanced_accuracy']

# accuracy
acc_results[model] = ConvTran_model_evals['model_accuracy']

# macro precision
macro_preci_results[model] = ConvTran_model_evals['macro_precision']

# macro sensitivity & specificity
macro_sensitivity_results[model], macro_specificity_results[model] = calculate_macro_metrics(ConvTran_evals['confusion_matirx'])

# macro f1-score
macro_f1_score_results[model] = ConvTran_model_evals['macro_f1-score']

# -------------------------- Deep Learning Results --------------------------

# Deep Learning Results
for dl_model, all_evals in dl_results_dict.items():
    model_eval = all_evals['additional_metrics']
    
    # balanced accuracy
    ba_results[dl_model] = model_eval['balanced_accuracy']
    
    # accuracy
    acc_results[dl_model] = model_eval['model_accuracy']
    
    # macro precision
    macro_preci_results[dl_model] = model_eval['macro_precision']
    
    # macro sensitivity & specificity
    macro_sensitivity_results[dl_model], macro_specificity_results[dl_model] = calculate_macro_metrics(all_evals['confusion_matirx'])

    # macro f1-score
    macro_f1_score_results[dl_model] = model_eval['macro_f1-score']
```

```python
def appoint_colors(keys):
    colors = []
    for key in keys:
        if key == 'ROCKET':
            colors.append('green')
        elif key == 'ConvTran':
            colors.append('orange')
        else:
            colors.append('blue')
    return colors
```

```python
def change_names(keys):
    new_keys = []
    for key in keys:
        if key == 'inception':
            new_keys.append('Inception')
        elif key == 'resnet':
            new_keys.append('RESNET')
        elif key == 'encoder':
            new_keys.append('Encoder')
        elif key == 'mlp':
            new_keys.append('MLP')
        elif key == 'mcdcnn':
            new_keys.append('MCDCNN')
        elif key == 'fcn':
            new_keys.append('FCN')
        elif key == 'cnn':
            new_keys.append('Time-CNN')
        elif key == 'twiesen':
            new_keys.append('TWIESN')
        elif key =='mcnn':
            new_keys.append('MCNN')
        elif key == 'tlenet':
            new_keys.append('t-LeNet')
        else:
            new_keys.append(key)
    return new_keys
```

# Confusion matrices

```python
model_names = ['ROCKET', 'ConvTran', 'Inception', 'Encoder']
```

```python
def norm_cm(cm):
    return cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
```

```python
confusion_matrices = []

confusion_matrices.append(norm_cm(ROCKET_evals['confusion_matirx']))
confusion_matrices.append(norm_cm(ConvTran_evals['confusion_matirx']))
confusion_matrices.append(norm_cm(dl_results_dict['inception']['confusion_matirx']))
# confusion_matrices.append(norm_cm(dl_results_dict['resnet']['confusion_matirx']))
confusion_matrices.append(norm_cm(dl_results_dict['encoder']['confusion_matirx']))
# confusion_matrices.append(norm_cm(dl_results_dict['mlp']['confusion_matirx']))
# confusion_matrices.append(norm_cm(dl_results_dict['mcdcnn']['confusion_matirx']))
```

```python
fig, axes = plt.subplots(2, 2, figsize=(15, 20))

# Increase the vertical space between subplots
plt.subplots_adjust(hspace=0.1, wspace=0.4)

# Fill in only 7 subplots, leave the last one empty
for i, ax in enumerate(axes.flat):
    if i < 7:
        sns.heatmap(confusion_matrices[i], annot=True, fmt=".2f", ax=ax,
                    xticklabels=class_labels, yticklabels=class_labels, cmap='cividis',
                    annot_kws={"size": 10}, cbar=False, square=True)
        
        ax.set_title(str(model_names[i]), fontsize=18)  # Increase title font size
        ax.set_xlabel('Predicted', fontsize=16)
        ax.set_ylabel('Actual', fontsize=16)
        ax.tick_params(axis='both', which='major', labelsize=14)
        plt.setp(ax.get_xticklabels(), rotation=90)  # Rotate x-axis tick labels by 45 degrees
        plt.setp(ax.get_yticklabels(), rotation=45)  # Rotate x-axis tick labels by 45 degrees
    else:
        ax.axis('off')  # Turn off the remaining empty subplot

# Adjust the layout to make room for the color bar
plt.subplots_adjust(right=0.85)

# Add a color bar
cbar_ax = fig.add_axes([0.88, 0.15, 0.02, 0.7])
fig.colorbar(plt.cm.ScalarMappable(cmap='cividis'), cax=cbar_ax)

plt.show()
```

```python
# Ensure each model's data has all class labels
for model in class_wise_results_specificity.keys():
    for label in class_labels:
        if label not in class_wise_results_specificity[model]:
            class_wise_results_specificity[model][label] = default_value

# Correct number of ends for each radar plot
num_classes = len(class_labels)

fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(18, 18), subplot_kw=dict(polar=True))
axes = axes.flatten()

# Increase font sizes
# plt.rcParams.update({'font.size': 12})

for idx, model in enumerate(models):
    values = [class_wise_results_specificity[model][label] for label in class_labels]
    values += values[:1]  # Closing the circle
    angles = [n / float(num_classes) * 2 * np.pi for n in range(num_classes)]
    angles += angles[:1]

    ax = axes[idx]
    ax.set_theta_offset(np.pi / 2)
    ax.set_theta_direction(-1)
    ax.set_rlabel_position(0)
    ax.set_xticks(angles[:-1])
    ax.set_xticklabels(class_labels, fontsize=18)
    ax.plot(angles, values, linewidth=2, linestyle='solid', label=model)
    ax.fill(angles, values, alpha=0.4)
    ax.set_title(model, fontsize=20)

# Hide any unused subplots
for ax in axes[7:]:
    ax.set_visible(False)

# Add a main title
fig.suptitle('Class-wise Specificity for ROCKET, ConvTran and the Best DL Models', fontsize=24)

plt.tight_layout(pad=3.0)
plt.show()
```
